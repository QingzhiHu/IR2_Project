```python
data1 = pd.read_csv("docleaderboard-queries.tsv", sep='\t', header=None)
data1.rename({0: 'qid', 1: 'query'}, axis=1, inplace=True)
data1
```

```python
data2 = pd.read_csv("docleaderboard-top100.tsv", sep=' ', header=None)
data2.rename({0: 'qid', 1: 'Q0', 2:'docid', 3:'rank', 4:'score', 5:'runstring'}, axis=1, inplace=True)
data2
```

```python
temp = pd.merge(data1, data2, on='qid')[["qid","query","docid","rank"]]
temp
```

```python
temp[temp["qid"]<50000]
```

```python
# df = temp[temp["rank"]==1]
# df = df.drop(columns=['rank'])
# df
df = temp[temp["qid"]<50000]
```

```python
docid_list = df[["docid"]].values.tolist()
docid_list = [item for sublist in docid_list for item in sublist]
docid_list
```

```python
from tqdm import tqdm
# Opening file
with open('msmarco_doc_merge.tsv',"r",encoding="utf8") as in_file:
    with open('msmarco_doc_merge_processed_test5.txt',"w",encoding="utf8") as out_file:
        for line in tqdm(in_file):
            ls = line.split("\t")
            if ls[0] in docid_list:
                out_file.write(ls[0]+"\t"+ls[1]+" "+ls[2])
                    
```

```python
data3 = pd.read_csv('msmarco_doc_merge_processed_test5.txt', sep='\t', header=None)
data3
```

```python
data3.rename({0: 'docid', 1: 'document'}, axis=1, inplace=True)
data3
```

```python
df_final = pd.merge(df, data3, on='docid')[["qid","docid","query","document"]]
df_final
```

```python
df_final.to_csv('msmarco_doc_test5.tsv', header=None, index=None, sep='\t', mode='a')
```
