```python
import ir_datasets
dataset = ir_datasets.load("vaswani")
```

```python
# import ir_datasets
# dataset = ir_datasets.load("trec-robust04")
```

```python
import pandas as pd
query_ids = []
queries = []
for query in dataset.queries_iter():
    query_ids.append(query[0])
    queries.append(query[1])
df1 = pd.DataFrame()
df1["qid"] = query_ids
df1["query"] = queries
df1
#     print(query[0]) # namedtuple<query_id, title, description, narrative>
```

```python
doc_ids = []
docs = []
for doc in dataset.docs_iter():
    doc_ids.append(doc[0])
    docs.append(doc[1])
df2 = pd.DataFrame()
df2["docid"] = doc_ids
df2["document"] = docs
df2
```

```python
qrels_query_ids = []
qrels_doc_ids = []
qrels_relevance = []
for qrel in dataset.qrels_iter():
    qrels_query_ids.append(qrel[0])
    qrels_doc_ids.append(qrel[1])
    qrels_relevance.append(qrel[2])
    
df3 = pd.DataFrame()
df3["qid"] = qrels_query_ids
df3["docid"] = qrels_doc_ids
df3["relevance"] = qrels_relevance
df3
```

```python
df_qrels = df3.copy()
df_qrels['dummy'] = 0
df_qrels[["qid","dummy","docid","relevance"]].to_csv('vaswani_qrels.tsv', header=None, index=None, sep='\t', mode='a')
```
